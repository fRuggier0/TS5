```python
import scipy.signal as sig
from pytc2.sistemas_lineales import pretty_print_lti, tf2sos_analog, pretty_print_SOS
from pytc2.general import print_latex, print_subtitle
from pytc2.sistemas_lineales import analyze_sys, bodePlot, pzmap


n = 3

z,p,k = sig.besselap(n, norm='delay')

num_bessel, den_bessel = sig.zpk2tf(z,p,k)

sos_bessel = tf2sos_analog(num_bessel, den_bessel)

Taux = sig.TransferFunction(num_bessel , den_bessel)

pretty_print_lti(Taux)
print(' ')
pretty_print_SOS(sos_bessel)
print(' ')
pretty_print_SOS(sos_bessel, mode='omegayq')
print(' ')

all_sys = []

all_sys.append( sig.TransferFunction(num_bessel, den_bessel) )

figaxes = analyze_sys( all_sys )
```
